```python
import pandas as pd
```

```python
def aaMap(_aa):
    aa_df = pd.DataFrame([['Alanine','Ala','A'],
                        ['Arginine','Arg','R'],
                        ['Asparagine','Asn','N'],
                        ['Aspartate','Asp','D'],
                        ['Cysteine','Cys','C'],
                        ['Glutamate','Glu','E'],
                        ['Glutamine','Gln','Q'],
                        ['Glycine','Gly','G'],
                        ['Histidine','His','H'],
                        ['Histidine_D','HID','H'],
                        ['Histidine_E','HIE','H'],
                        ['Histidine_P','HIP','H'],
                        ['Isoleucine','Ile','I'],
                        ['Leucine','Leu','L'],
                        ['Lysine','Lys','K'],
                        ['Methionine','Met','M'],
                        ['Phenylalanine','Phe','F'],
                        ['Proline','Pro','P'],
                        ['Serine','Ser','S'],
                        ['Threonine','Thr','T'],
                        ['Tryptophan','Trp','W'],
                        ['Tyrosine','Tyr','Y'],
                        ['Valine','Val','V']],columns=['Full','3','1'])
    aa_df['3'] = aa_df['3'].str.upper()
    _aa = _aa.upper()
    if len(_aa)==1:
        toret = aa_df.loc[aa_df['1']==_aa,'3'].values
        if len(toret)>0:
            toret = toret[0]
        else:
            return None
    else:
        toret = aa_df.loc[aa_df['3']==_aa,'1'].values
        if len(toret)>0:
            toret = toret[0]
        else:
            return None
    return toret
```

```python
from biopandas.pdb import PandasPdb
ppdb = PandasPdb().read_pdb('2otu.pdb')
atom_df = ppdb.df['ATOM'].copy(deep=True)
atom_df
```

```python
structure = parser.get_structure('2otu', 'ot/2otu.cif') ## give the structure a name and path of file
```

```python
## Let's check the attribute in the file
def cleandir(obj):
    print(", ".join([a for a in dir(obj)]))
cleandir(structure)
```

```python
## Get rid of attributes starting with '_'
def cleandir(obj):
    print(", ".join([a for a in dir(obj) if not a.startswith("_")]))
cleandir(structure)
```

model in the structure contains several "chains" or strands of protein/RNA/DNA/.

Each "chain" is made up of residues, or amino acids/DNA bases/RNA bases.

Each residue is made up of Atoms.

```python
## extract the chain_id from the model
model = structure[0] #since we have one model (the structure can have multiple models)
for chain in model:
    print(f"chain {chain}, Chain ID: {chain.id}")
```

```python
## Let's access "A" chain
chain_A = model['A']
chain_A
len(chain_A)
```

```python
for res in chain_A:
    print("Residue name: ",res.resname, "number:",res.id[1]) ##res.id is a tuple of 3 entries, get the middle one
```
