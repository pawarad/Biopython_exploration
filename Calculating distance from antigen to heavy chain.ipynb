```python
import numpy as np
import pandas as pd
from biopandas.pdb import PandasPdb
```

```python
ppdb = PandasPdb().read_pdb('4ydl.pdb')
atom_df = ppdb.df['ATOM'].copy(deep=True)
atom_df.head()
```

```python
Heavy_chain = atom_df[atom_df['chain_id'] == 'B'] ## Heavy chain
# Heavy_chain.drop_duplicates(['chain_id','residue_number'])
Heavy_chain
```

```python
Light_chain = atom_df[atom_df['chain_id'] == 'C'] ## Light chain
# Light_chain.drop_duplicates(['chain_id','residue_number']).head()
Light_chain
```

```python
Antigen_chain = atom_df[atom_df['chain_id'] == 'A'] ## Antigen chain
Antigen_chain
```

```python
A_coord = Antigen_chain[['x_coord','y_coord','z_coord']].to_numpy()
H_coord = Heavy_chain[['x_coord','y_coord','z_coord']].to_numpy()
L_coord = Light_chain[['x_coord','y_coord','z_coord']].to_numpy()
```

```python
## Looping every residue from heavy chain overy every residue from antigen chain to calculate distance 

distance = []
for a in range(len(A_coord)):
    dep=[]
    for h in range(len(H_coord)):
        dist = np.sqrt(int(A_coord[a][0]-H_coord[h][0])^2 +
                       int(A_coord[a][1]-H_coord[h][1])^2+
                       int(A_coord[a][2]-H_coord[h][2])^2)
        dep.append(dist)
    distance.append(dep)
```

```python
dist_matrix = np.array(distance)
dist_matrix.shape  
```

```python
row_names = 'Antigen_' + Antigen_chain['residue_name'].to_numpy()
column_names = 'H_' + Heavy_chain['residue_name'].to_numpy()
```

```python
df_heavy = pd.DataFrame(dist_matrix, columns=column_names, index=row_names)
```

### Light Chain distance

```python
## Looping every residue from heavy chain overy every residue from antigen chain to calculate distance 

distance_L = []
for a in range(len(A_coord)):
    dep=[]
    for h in range(len(L_coord)):
        dist = np.sqrt(int(A_coord[a][0]-L_coord[h][0])^2 +
                       int(A_coord[a][1]-L_coord[h][1])^2+
                       int(A_coord[a][2]-L_coord[h][2])^2)
        dep.append(dist)
    distance_L.append(dep)
```

```python
L_dist_matrix = np.array(distance_L)
L_dist_matrix.shape  
```

```python
row_names = 'Antigen_' + Antigen_chain['residue_name'].to_numpy()
column_names = 'L_' + Light_chain['residue_name'].to_numpy()
df_light = pd.DataFrame(L_dist_matrix, columns=column_names, index=row_names)
```
